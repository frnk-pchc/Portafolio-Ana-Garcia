###Ingresos por tienda

````python
ingreso_tienda1=tienda['Precio'].sum()
ingreso_tienda2=tienda2['Precio'].sum()
ingreso_tienda3=tienda3['Precio'].sum()
ingreso_tienda4=tienda4['Precio'].sum()
ingresos_df=pd.DataFrame({"Tiendas":["Tienda 1",
                                     "Tienda 2",
                                     "Tienda 3",
                                     "Tienda 4"],
                          "Total":[ingreso_tienda1,ingreso_tienda2,ingreso_tienda3,ingreso_tienda4]})
display(ingresos_df)
````

Gráfica de Barra para comparar y visualizar de mejor manera los ingresos en cada tienda.

````python

plt.figure(figsize=(10, 6))
bars = plt.bar(ingresos_df["Tiendas"], ingresos_df["Total"], color="#1f77b4", edgecolor='black', alpha=0.85)

plt.title("Ingresos por Tienda", fontsize=16, fontweight='bold')
plt.xlabel("Tiendas", fontsize=14)
plt.ylabel("Ingresos ($)", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Añadir etiquetas de valor encima de cada barra
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:,.0f}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 5),  # Desplazamiento vertical de la etiqueta
                 textcoords="offset points",
                 ha='center', va='bottom', fontsize=12)

plt.tight_layout()
plt.show()
````

Gráfico Circular

````python
import matplotlib.pyplot as plt

# Datos de las tiendas
labels = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
sizes = [1150880400, 1116343500, 1098019600, 1038375700]
colors = ['#4daf4a', '#377eb8', '#ff7f00', '#f781bf']

def func(pct, allvals):
    absolute = int(pct/100.*sum(allvals))
    return f"{labels[func.idx]}: {pct:.1f}%"
func.idx = -1

fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(aspect="equal"))

# Función para recorrer índices en el formateo de etiquetas
def autopct_generator(allvals):
    def inner(pct):
        inner.index = (inner.index + 1) if hasattr(inner, "index") else 0
        func.idx = inner.index
        return func(pct, allvals)
    return inner

wedges, texts, autotexts = ax.pie(
    sizes,
    autopct=autopct_generator(sizes),
    textprops=dict(color="black", fontsize=12, weight='bold'),
    colors=colors,
    startangle=140,
    wedgeprops=dict(width=0.4, edgecolor='w')
)

ax.legend(wedges, labels, title="Tiendas", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), fontsize=12)

plt.setp(autotexts, size=12, weight="bold")
ax.set_title("Distribución de Ingresos Totales por Tienda", fontsize=16, weight='bold')

plt.show()
````

Reporte de Calificacion por Tiendas

````python
import pandas as pd

def calcular_calificacion(lista_tiendas):
    calificaciones = []

    for id_tienda, df_tienda in lista_tiendas:
        promedio = df_tienda['Calificación'].mean().round(2)
        nombre = f"Tienda {id_tienda}"
        calificaciones.append({'Tienda': nombre, 'Calificación Promedio': promedio})
        print(f"La calificación promedio de {nombre} es: {promedio:.2f}")

    # Convertir lista de resultados a DataFrame para mejor manejo y visualización
    calificaciones_df = pd.DataFrame(calificaciones)
    return calificaciones_df

# Uso de la función y guardar resultados en DataFrame
calificaciones_df = calcular_calificacion(lista_tiendas)
````

###

````python
import matplotlib.pyplot as plt

def grafica_calificaciones(calificaciones_df):
    tiendas_label = calificaciones_df['Tienda']
    calificaciones_label = calificaciones_df['Calificación Promedio']

    plt.figure(figsize=(8, 5))
    plt.scatter(tiendas_label, calificaciones_label, color='#1f77b4', s=150, alpha=0.8, edgecolor='black', linewidth=1)

    # Añadir etiquetas con valores encima de cada punto
    for i, valor in enumerate(calificaciones_label):
        plt.text(i, valor + 0.02, f"{valor:.2f}", ha='center', va='bottom', fontsize=11, fontweight='bold')

    plt.title('Calificación Promedio por Tienda', fontsize=16, fontweight='bold')
    plt.xlabel('Tiendas Alura Store', fontsize=14)
    plt.ylabel('Calificación', fontsize=14)
    plt.ylim(0, 5)  # Asumiendo que la calificación es sobre 5
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

grafica_calificaciones(calificaciones_df)
````

Más vendido y menos vendidos

````python
import matplotlib.pyplot as plt

tiendas = {"Tienda 1": tienda, "Tienda 2": tienda2, "Tienda 3": tienda3, "Tienda 4": tienda4}

for nombre, df in tiendas.items():
    # Conteo total de ventas por producto en la tienda
    ventas_producto = df['Producto'].value_counts()

    # Extraer top 5 más vendidos y top 5 menos vendidos
    mas_vendidos = ventas_producto.head(5)
    menos_vendidos = ventas_producto.tail(5)

    print(f"\nTop 5 productos más vendidos en {nombre}:\n")
    display(mas_vendidos)

    print(f"\nTop 5 productos menos vendidos en {nombre}:\n")
    display(menos_vendidos)

    # Graficar top 5 más vendidos
    plt.figure(figsize=(10, 4))

    plt.subplot(1, 2, 1)
    mas_vendidos.plot(kind='bar', color='#ff7f0e', edgecolor='black', alpha=0.85)
    plt.title(f'Top 5 productos más vendidos - {nombre}')
    plt.xlabel('Producto')
    plt.ylabel('Ventas')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Graficar top 5 menos vendidos
    plt.subplot(1, 2, 2)
    menos_vendidos.plot(kind='bar', color='#1f77b4', edgecolor='black', alpha=0.85)
    plt.title(f'Top 5 productos menos vendidos - {nombre}')
    plt.xlabel('Producto')
    plt.ylabel('Ventas')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.show()
````

````python
import matplotlib.pyplot as plt

tiendas = {"Tienda 1": tienda, "Tienda 2": tienda2, "Tienda 3": tienda3, "Tienda 4": tienda4}
mas_vendidos = {}
menos_vendidos = {}

for nombre, df in tiendas.items():
    # Calcular ventas por producto
    conteo_productos = df['Producto'].value_counts()
    mas_vendidos[nombre] = conteo_productos.head(5)
    menos_vendidos[nombre] = conteo_productos.tail(5)
    print(f"\nTop 5 productos más vendidos en {nombre}:")
    display(mas_vendidos[nombre])
    print(f"\nTop 5 productos menos vendidos en {nombre}:")
    display(menos_vendidos[nombre])
    # Gráficos de más vendidos
    plt.figure(figsize=(8, 4))
    mas_vendidos[nombre].plot(kind='bar', color='#ff7f0e', edgecolor='black', alpha=0.85)
    plt.title(f'Top 5 productos más vendidos - {nombre}', fontsize=14, fontweight='bold')
    plt.xlabel('Producto', fontsize=12)
    plt.ylabel('Cantidad de Ventas', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

    # Gráficos de menos vendidos
    plt.figure(figsize=(8, 4))
    menos_vendidos[nombre].plot(kind='bar', color='#1f77b4', edgecolor='black', alpha=0.85)
    plt.title(f'Top 5 productos menos vendidos - {nombre}', fontsize=14, fontweight='bold')
    plt.xlabel('Producto', fontsize=12)
    plt.ylabel('Cantidad de Ventas', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
````

````python
import pandas as pd
from IPython.display import display

def calculo_costo(lista_tiendas):
    costos_lista = []
    for id_tienda, df_tienda in lista_tiendas:
        costo_envio = df_tienda["Costo de envío"].mean().round(2)
        tienda = f"Tienda {id_tienda}"
        costos_lista.append({'Tienda': tienda, 'Costo promedio': costo_envio})
    costos_df = pd.DataFrame(costos_lista)
    display(costos_df)
    return costos_lista

costos_lista = calculo_costo(lista_tiendas)
````

````python
import matplotlib.pyplot as plt

def graficar_torta_envio(costos_lista):
    tiendas_nombres = [i['Tienda'] for i in costos_lista]
    costos_promedio = [i['Costo promedio'] for i in costos_lista]
    colores = ['#ff9999','#66b3ff','#99ff99','#ffd966']  # Colores suaves personalizados
    explode = (0.05, 0.05, 0.05, 0.05)  # Separar cada segmento ligeramente

    plt.figure(figsize=(8, 6))
    patches, texts, autotexts = plt.pie(costos_promedio,
                                        labels=tiendas_nombres,
                                        autopct='%1.1f%%',
                                        startangle=140,
                                        colors=colores,
                                        explode=explode,
                                        shadow=True,
                                        textprops={'fontsize': 12, 'weight': 'bold', 'color': 'black'})

    plt.title('Comparación de costos de envío promedio por tienda', fontsize=16, weight='bold')
    plt.legend(patches, tiendas_nombres, title='Tiendas', loc='best')
    plt.tight_layout()
    plt.show()

# Llama a la función con tu lista de costos
graficar_torta_envio(costos_lista)
````
